```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
test = pd.read_csv('test.csv')
```

```python
test.head()
```

# Feature Scaling and Mean Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(train.drop('label', axis=1))
```

```python
scaled_features = scaler.transform(train.drop('label', axis=1))
```

```python
scaled_test_features = scaler.transform(test)
```

```python
scaled_features
```

```python
df_feat = pd.DataFrame(scaled_features, columns=train.columns[1:])
```

```python
df_feat_test = pd.DataFrame(scaled_test_features, columns=test.columns)
```

```python
df_feat.head()
```

```python
df_feat_test.head()
```

# Reducing Feature Size using PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=100)
```

```python
pca.fit(df_feat)
```

```python
x_pca = pca.transform(df_feat)
```

```python
x_pca.shape
```

```python
df_feat.shape
```

# Splitting data into training and CV sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = x_pca
y = train['label']
```

```python
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.3)
```

# Random Forest Estimator

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X=X_train, y=y_train)
```

```python
preds = rfc.predict(X_cv)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_cv, preds))
print('\n')
print(classification_report(y_cv, preds))
```

# Variable Feature Size for PCA

```python
feat_size = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]

for size in feat_size:
    pca = PCA(n_components=size)
    pca.fit(df_feat)
    x_pca = pca.transform(df_feat)
    X = x_pca
    y = train['label']
    X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.3)
    rfc = RandomForestClassifier(n_estimators=100)
    rfc.fit(X=X_train, y=y_train)
    preds = rfc.predict(X_cv)
    print(classification_report(y_cv, preds))
    print('\n\n')
```

```python
feat_size = [10, 25, 50]

for size in feat_size:
    pca = PCA(n_components=size)
    pca.fit(df_feat)
    x_pca = pca.transform(df_feat)
    X = x_pca
    y = train['label']
    X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.3)
    rfc = RandomForestClassifier(n_estimators=100)
    rfc.fit(X=X_train, y=y_train)
    preds = rfc.predict(X_cv)
    print(classification_report(y_cv, preds))
    print('\n\n')
```

```python
feat_size = [75]

for size in feat_size:
    pca = PCA(n_components=size)
    pca.fit(df_feat)
    x_pca = pca.transform(df_feat)
    X = x_pca
    y = train['label']
    X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.3)
    rfc = RandomForestClassifier(n_estimators=100)
    rfc.fit(X=X_train, y=y_train)
    preds = rfc.predict(X_cv)
    print(classification_report(y_cv, preds))
    print('\n\n')
```
